```python
from torch.utils.data import DataLoader
from torch.utils.data import TensorDataset
import torch.nn as nn
import numpy as np
import torch
```

```python
def gen_mat(dim):
    matrix = np.zeros((dim, dim))
    for i in range(1, dim-1):
        matrix[i][i-1], matrix[i][i], matrix[i][i+1] = 1, -2, 1
    matrix[0][0], matrix[0][1] = -2, 1
    matrix[-1][-1], matrix[-1][-2] = -2, 1
    return matrix
```

```python
def finite_diff_method(lamda, x):
    h = 0.001
    dim = len(x) - 2
    mat = gen_mat(dim)
    xx = x[1:-1]
    C = -lamda * np.cos(np.pi * xx) * (h**2)
    C[0] = C[0] - 1
    C[-1] = C[-1] + 1
    u_np = np.linalg.inv(mat).dot(C) 
    u_np = np.append(u_np, -1)
    u_np = np.insert(u_np, 0, 1)
    return u_np
```

```python
# 准备数据
sample_num = 1001
x = np.linspace(0,1,sample_num)
y = finite_diff_method(np.pi**2, x)
# 将数据做成数据集的模样
X = np.expand_dims(x,axis=1)
Y = y.reshape(sample_num, -1)
X = torch.tensor(X, dtype = torch.float, requires_grad = True)
Y = torch.tensor(Y, dtype = torch.float, requires_grad = True)
```

```python
torch.set_printoptions(precision=8)
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.net=nn.Sequential(
            nn.Linear(in_features=1,out_features=10),nn.ReLU(),
            nn.Linear(10,100),nn.ReLU(),
            nn.Linear(100,10),nn.ReLU(),
            nn.Linear(10,1)
        )
    def forward(self, input:torch.FloatTensor):
        return self.net(input)

net=Net()

# set optimizer and lamda 
optim=torch.optim.Adam(Net.parameters(net),lr=0.001)
lamda = torch.randn((), dtype = torch.float, requires_grad=True)
Loss=nn.MSELoss()
```

```python
# 下面开始训练：
# 一共训练500次
for i in range(11):
    alpha = (i) * 0.1
    for epoch in range(1000):
        loss = None 

        # ----------l1------------
        y_predict = net(X)
        loss_1 = Loss(y_predict,Y)

        # ----------l2------------
        I1 = torch.ones_like(X)
        I1.requires_grad = True
        Ux = torch.autograd.grad(y_predict, X, grad_outputs=I1, create_graph = True)[0]
        Uxx = torch.autograd.grad(Ux.sum(), I1, create_graph = True)[0]
        Uxx_mean = torch.mean(Uxx)
        f = lamda * torch.cos(X * np.pi) + Uxx
    
        loss_2 = alpha * torch.mean(f**2)
        
        loss = loss_1 + loss_2

        optim.zero_grad()
        loss.backward()
        optim.step()
        with torch.no_grad():
            lamda -= 0.001 * lamda.grad
            lamda.grad = None
            

        # 每100次 的时候打印一次日志
        if (epoch+1)%100==0:
            print("step: {0}, alpha: {1}, loss: {2}, lamda: {3}, Uxx_mean{4}".format(epoch+1, alpha, loss.item(), lamda.item(), Uxx_mean))


# 使用训练好的模型进行预测
predict=net(X)

# 绘图展示预测的和真实数据之间的差异
import matplotlib.pyplot as plt
plt.plot(x,y,label="fact")
plt.plot(x,predict.detach().numpy(),label="predict")
plt.title("cos function")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.savefig(fname="result.png",figsize=[10,10])
plt.show()
```
